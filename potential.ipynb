```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
V0 = 0.5
a = 1 
N = 10
x =np.linspace(0,1,N+a)
y = np.linspace(0,1,N+a)
dx = x[1] - x[0]
dy = y[1] - y[0]
X, Y = np.meshgrid(x,y)
V = np.zeros([N+1, N+1])
V2 = np.zeros([N+1, N+1])
V[:,0] = V0
V2[:,0] = V0
Ex = 0*X
Ey = 0*Y
# print(V)

n = 0
NN = 100

while n < NN:
    for i in range(1,N):
        for j in range(1,N):
            V2[i,j] = (V[i+1,j] + V[i-1,j] + V[i, j+1] + V[i, j-1]) / 4
    for i in range(1,N):
        for j in range(1,N):
            V[i,j] = V2[i,j]
    n = n + 1

for i in range(1,N):
        for j in range(1,N):
             Ex[i,j] = -(V[i+1,j] - V[i-1,j]) / (2*dx)
             Ey[i,j] = -(V[i,j+1] - V[i,j-1]) / (2*dy)


fig, ax = plt.subplots()
contour = ax.contourf(X,Y,V, levels=20)
ax.quiver(X,Y,Ey,Ex,color='white')

# plt.quiver(X,Y,Ey,Ex)
# plt.contourf(X,Y,V, levels=20)
# fig = plt.figure()
# ax = fig.add_subplot(111, projection='3d')
# ax.plot_surface(X,Y,V)
```
